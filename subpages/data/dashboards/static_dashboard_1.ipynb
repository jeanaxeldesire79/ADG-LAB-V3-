# Static dashboard exporter

Use this notebook to preview the West Africa Human Development dashboard. The exported HTML calls the UNDP Human Development Report API (with an automatic Our World in Data fallback) and renders the analytics with Plotly in the browser.

## Usage

1. Run the cell below to read the packaged HTML.
2. The HTML is stored at `subpages/data/dashboards/static_dashboard_1.html` ready for embedding.
3. Set your HDR API key once in the browser (e.g. `localStorage.setItem('hdr_api_key', 'YOUR_KEY')`) before loading the page.
4. When embedding elsewhere, keep the `<script>` tags so the live data fetch continues to work.

```python
from pathlib import Path
from IPython.display import HTML

dashboard_path = Path("subpages/data/dashboards/static_dashboard_1.html")
dashboard_html = dashboard_path.read_text(encoding="utf-8")
print(f"{dashboard_path} â†’ {len(dashboard_html):,} characters")

HTML(
    f"""
    <iframe
      src="{dashboard_path.name}"
      style="width:100%;max-width:100%;height:860px;border:1px solid rgba(148,163,184,0.35);border-radius:18px;background:white"
      loading="lazy"
      title="Static dashboard preview"
    ></iframe>
    """
)
```
